## Sales 

```python
# Sales Section 

#-------------------------------------------------------------------------------------#
# Create an end point for sales by transaction by day 
from datetime import datetime
start_date = "2016-09-01"
stop_date = "2018-04-28"

start = datetime.strptime(start_date, "%Y-%m-%d")
stop = datetime.strptime(stop_date, "%Y-%m-%d")
Flag = 0 
from datetime import timedelta
while start <= stop:
    startdate = str(start.date())
    end_point = "/sales/v1/transactions/"+ startdate + "?licenseNumber=" + license_no
    
    print(startdate)
    
    # Send a request to the url 
    r = requests.get(url+end_point , auth = HTTPBasicAuth(user_name,user_pass))
    
    print(r.status_code)
    
    # Save to file if response is 200
    if (r.status_code == 200):
        with open('data.json', 'w') as f:
          json.dump(r.json(), f, ensure_ascii=False)
        data = pd.read_json("data.json")
        
        if (Flag == 0 ):
             with open('metric_api_data/sales_by_day-' + license_no + '.csv','a') as g:
                data.to_csv(g)
                Flag = Flag + 1
        
        
        with open('metric_api_data/sales_by_day-' + license_no + '.csv','a') as g:
             data.to_csv(g, header=False)
        
       
    start = start + timedelta(days=1)
    
# Print the status of the response
print(r.status_code)
```
